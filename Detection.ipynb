exp 23 - plot

```python
import pandas as pd
import matplotlib.pyplot as plt

path_combined = '/content/drive/MyDrive/Detection/yolov5/runs/train/exp23/results.csv'

# Specify the path to your CSV file
csv_file_path = path_combined

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file_path)

# Remove leading spaces from column names
data.columns = data.columns.str.strip()

# Extract the columns
epochs = data['epoch']
train_box_loss = data['train/box_loss']
train_obj_loss = data['train/obj_loss']
train_cls_loss = data['train/cls_loss']
val_box_loss = data['val/box_loss']
val_obj_loss = data['val/obj_loss']
val_cls_loss = data['val/cls_loss']
precision = data['metrics/precision']
recall = data['metrics/recall']
mAP_05 = data['metrics/mAP_0.5']
mAP_0595 = data['metrics/mAP_0.5:0.95']

# Create a figure with subplots
fig, axs = plt.subplots(2, 5, figsize=(20, 8))

plt.style.use('ggplot')

# Plot train/box_loss
axs[0, 0].plot(epochs, train_box_loss)
axs[0, 0].set_xlabel('Epochs')
axs[0, 0].set_ylabel('Loss')
axs[0, 0].set_title('train/box_loss')

# Plot train/obj_loss
axs[0, 1].plot(epochs, train_obj_loss)
axs[0, 1].set_xlabel('Epochs')
axs[0, 1].set_ylabel('Loss')
axs[0, 1].set_title('train/obj_loss')

# Plot train/cls_loss
axs[0, 2].plot(epochs, train_cls_loss)
axs[0, 2].set_xlabel('Epochs')
axs[0, 2].set_ylabel('Loss')
axs[0, 2].set_title('train/cls_loss')

# Plot metrics/precision
axs[0, 3].plot(epochs, precision)
axs[0, 3].set_xlabel('Epochs')
axs[0, 3].set_ylabel('Precision')
axs[0, 3].set_title('metrics/precision')

# Plot metrics/recall
axs[0, 4].plot(epochs, recall)
axs[0, 4].set_xlabel('Epochs')
axs[0, 4].set_ylabel('Recall')
axs[0, 4].set_title('metrics/recall')

# Plot val/box_loss
axs[1, 0].plot(epochs, val_box_loss)
axs[1, 0].set_xlabel('Epochs')
axs[1, 0].set_ylabel('Loss')
axs[1, 0].set_title('val/box_loss')

# Plot val/obj_loss
axs[1, 1].plot(epochs, val_obj_loss)
axs[1, 1].set_xlabel('Epochs')
axs[1, 1].set_ylabel('Loss')
axs[1, 1].set_title('val/obj_loss')

# Plot val/cls_loss
axs[1, 2].plot(epochs, val_cls_loss)
axs[1, 2].set_xlabel('Epochs')
axs[1, 2].set_ylabel('Loss')
axs[1, 2].set_title('val/cls_loss')

# Plot metrics/mAP_0.5
axs[1, 3].plot(epochs, mAP_05)
axs[1, 3].set_xlabel('Epochs')
axs[1, 3].set_ylabel('mAP')
axs[1, 3].set_title('metrics/mAP_0.5')

# Plot metrics/mAP_0.5:0.95
axs[1, 4].plot(epochs, mAP_0595)
axs[1, 4].set_xlabel('Epochs')
axs[1, 4].set_ylabel('mAP')
axs[1, 4].set_title('metrics/mAP_0.5:0.95')

# Adjust the spacing between subplots
plt.tight_layout()

# Save the figure as an image file
plt.savefig('all_plots-exp23.png')
plt.show()
```
